```python
import numpy as np
import pandas as pd
import math

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_chang_rates(K, T):
    R = (K * K) / (K + T) ** 2
    return R
```

```python
K, L = 20, 20
T = np.arange(1, 60)
chang_rate = get_chang_rates(K, T)

plt.plot(T, chang_rate, c='r')
plt.title("Scheme by Chang for K=L=20")
plt.xlabel("T = security level")
plt.ylabel("R = download rate")
plt.show()
```

```python
def get_kakar_rates(K, L, T):
    R = (K * L) / ((K + T) * (L + 1) - 1)
    return R
```

```python
K, L = 20, 20
T = np.arange(1, 60)
kakar_rate = get_kakar_rates(K, L, T)

plt.plot(T, kakar_rate, c='r')
plt.title("Scheme by Chang for K=L=20")
plt.xlabel("T = security level")
plt.ylabel("R = download rate")
plt.show()
```

# Let's look through the comparison of the download rates:

```python
plt.plot(T, chang_rate, c='g')
plt.plot(T, kakar_rate, c='r')
plt.title("Schemes comparison for K=L=20")
plt.xlabel("T = security level")
plt.ylabel("R = download rate")
plt.gca().legend(["Chang's scheme", "Kakar's scheme"])
plt.show()
```

```python
def get_yang_rates(K, L):
    R = np.zeros([len(K), len(L)])
    for i in range(len(K)):
        for j in range(len(L)):
            R[i][j] = (K[i] * L[j]) / (K[i] * L[j] + K[i] + L[j])
    return R
```

```python
K, L = np.arange(1, 20), np.arange(1, 20)
yang_rate = get_yang_rates(K, L)
print(yang_rate.shape)
plt.imshow(yang_rate, cmap='hot', interpolation='nearest')
plt.title("Scheme by Yang for T=1")
plt.xlabel("dimension K")
plt.ylabel("dimension L")
plt.show()
```
